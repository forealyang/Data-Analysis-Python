2. Create price_label column 

```python
#define function 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply function 
df['price_range'] = df.apply(price_label, axis =1)
```

```python
#check values
df['price_range'].value_counts(dropna=False)
```

```python
#find most expensive product 
df['prices'].max()
```

```python
#using loc() function 
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
#use loc() on entire dataframe 
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

3. Create busiest_day columm

```python
#find which day most orders take place 
ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

0=Saturday has the most orders, 4=Wednesday has the lowest

```python
#for loop, new columns, busiest/least busy/regular busy 
result = []

for x in ords_prods_merge["orders_day_of_week"]:
  if x == 0:
    result.append("Busiest day")
  elif x == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#add result to ords_prods_merge 
ords_prods_merge['busiest day'] = result
```

```python
ords_prods_merge['busiest day'].value_counts(dropna=False)
```

4. Find busiest days


busiest days are: Saturday & Sunday
least busiest days are: Tuesday & Wednesday

```python
#new for loop

result2 = []

for value in ords_prods_merge['orders_day_of_week']:
  if value == 0 or value ==1:
    result2.append('Busiest days')
  elif value == 4 or value == 3:
    result2.append('Slowest days')
  else:
    result2.append('Regularly days')
```

```python
#check results
result2
```

```python
#create new column 'busiest_days'
ords_prods_merge['busiest_days']=result2
```

6.


When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#find frequency of each hour 
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

Most orders: 10, 11, 14, 15, 13, 12, 16, 9
Average orders: 17, 8, 18, 19, 20, 7, 21, 22
Fewest orders: 3, 4, 2, 5, 1, 0, 6, 23

```python
#create for loop 
result3 = []

for hour in ords_prods_merge['order_hour_of_day']:
  if hour in [9,10,11,12,13,14,15,16]:
    result3.append('Most orders')
  elif hour in [7,8,17,18,19,20,21,22]:
    result3.append('Average orders')
  else:
    result3.append('Fewest orders')
```

```python
#create new column 'busiest_period_of_day'
ords_prods_merge['busiest_period_of_day']=result3
```

```python
#return df to check new column 
ords_prods_merge
```
